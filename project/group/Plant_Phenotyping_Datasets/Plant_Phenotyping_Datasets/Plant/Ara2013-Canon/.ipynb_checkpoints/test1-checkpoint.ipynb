```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
img = cv2.imread("ara2013_plant051_rgb.png" )

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()

img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(img_gray, 'gray')
plt.show()
```

```python
def sift_features(img_list):
    des_list =[]
    sift = cv2.xfeatures2d.SIFT_create(200)
    for img in img_list:
        img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        kpG,  desG = sift.detectAndCompute(img_gray, None)
        # get features matrix of image
        des_list.append(desG)
    features =[]
    for kp_vector in des_list:
        for item in kp_vector:
            features.append(item)
    return features, des_list
```

```python
list_image =[]

list_image.append(img_rgb)
features, des_list  = sift_features(list_image)
```

```python
median_img41_rgb = cv2.medianBlur(img_rgb, 3)
plt.imshow(median_img41_rgb)
```

```python
kernel =np.ones((3,3),np.uint8)
erosion =cv2.erode(img_rgb,kernel,iterations =1)
opening = cv2.morphologyEx(median_img41_rgb, cv2.MORPH_OPEN, kernel)
plt.imshow(opening)
```

```python
mean_img_rgb = cv2.blur(img_rgb,(5,5)) 
plt.imshow(mean_img_rgb)
```

```python
img = mean_img_rgb
```

```python
# GMM
# grabcut 
mask = np.zeros(img.shape[:2], np.uint8)
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)

rect = (50,50,450,290)
cv2.grabCut(img,mask,rect,bgdModel,fgdModel, 5, cv2.GC_INIT_WITH_RECT )

mask2 = np.where((mask==2)|(mask==0), 0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]
plt.imshow(img),plt.colorbar(),plt.show()
```
