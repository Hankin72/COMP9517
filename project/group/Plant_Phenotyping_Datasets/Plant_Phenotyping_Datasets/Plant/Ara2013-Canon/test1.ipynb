```python
img = cv2.imread("ara2013_plant051_rgb.png" )

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.show()

img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(img_gray, 'gray')
plt.axis('off')
plt.show()

img_list = []

img_list.append(img_gray)
```

```python

sift = cv2.xfeatures2d.SIFT_create(200)
kps, des =cv2.xfeatures2d_SURF.detectAndCompute(sift, img_gray, None )
sift_img = cv2.drawKeypoints(img_gray,kps,np.array([]))
plt.axis('off')
plt.imshow(sift_img)
```

```python
features = []
reshape_features = []
sift = cv2.xfeatures2d.SIFT_create(200)
for img in img_list:
    _, desG = sift.detectAndCompute(img, None)
    features.append(desG)
    
for kp_vector in features:
    for item in kp_vector:
        reshape_features.append(item)
```

```python
plt.imshow(features[0])
```

```python
k=100
centroids, _ = kmeans(reshape_features, k, 1)
train_features_km = np.zeros((len(features), k), "float32")
    # print(train_features_km)
for index in range(len(features)):
        # labels --classify .. 为分类的标签
    labels, _ = vq(features[index], centroids)
    for i in labels:
        train_features_km[index][i] += 1

train_features_km
```
