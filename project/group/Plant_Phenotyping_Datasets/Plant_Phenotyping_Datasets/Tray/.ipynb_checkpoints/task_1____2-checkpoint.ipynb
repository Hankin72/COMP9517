```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from collections import Counter
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

from skimage import io,data,color

```

```python

# img = cv2.imread("./Ara2012/ara2012_tray02_rgb.png")
img = cv2.imread("./Ara2013-Canon/ara2013_tray02_rgb.png")
# img = cv2.imread("./Ara2013-RPi/ara2013_tray01_rgb.png")
# img = cv2.imread("./Ara2012/ara2012_tray02_rgb.png")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)


img_res = cv2.resize(img,(600,400),interpolation=cv2.INTER_CUBIC)
img_rgb_res = cv2.resize(img_rgb,(600,400),interpolation=cv2.INTER_CUBIC)
img_gray_res =  cv2.resize(img_gray,(600,400),interpolation=cv2.INTER_CUBIC)


# img_lab = cv2.cvtColor(img_res, cv2.COLOR_BGR2LAB)
plt.figure(figsize=(10,12))

plt.imshow(img_rgb_res)
```

```python
def a_b_tfb(img):
    i = img[:,:,1]
    j = img[:,:,0]
    f = cv2.blur(i, ksize=(3,3)) + (cv2.GaussianBlur(j,(3,3),4) - cv2.GaussianBlur(j,(3,3),1))
    tfb = np.exp(-np.abs(f)/50)
    return np.array([i,j,tfb]).transpose(1,2, 0), tfb

m,n,l = img_res.shape
lab = cv2.cvtColor(img_res, cv2.COLOR_BGR2LAB)

fs, tfb = a_b_tfb(lab)

X = tfb.reshape((-1,1))
y_hat = KMeans(n_clusters=2, random_state=0).fit_predict(X)
img_ = y_hat.reshape((m,n))

plt.imshow(img_)
plt.show()
```

```python
img_temp = cv2.cvtColor(green_1, cv2.COLOR_HSV2BGR)
img_hsv_2_gray = cv2.cvtColor(img_temp, cv2.COLOR_BGR2GRAY)


_, thresh = cv2.threshold(img_hsv_2_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# _, thresh = cv2.threshold(img_hsv_2_gray, 0, 255, cv2.THRESH_BINARY )

#mean_blur
thresh = cv2.blur(thresh,(2,2)) 
# img_morph = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
# img_morph = cv2.erode(img_morph, kernel, iterations=1)
#

img_dila = cv2.dilate(thresh, (3,3), iterations=7)

plt.imshow(img_hsv_2_gray, 'gray')
plt.show()

plt.imshow(img_dila,'gray')
plt.show()
```

```python
dst_1 = cv2.fastNlMeansDenoising(img_dila,None,10,7,21)
plt.imshow(dst_1)
# dst_dilat = dilatiion = cv2.dilate(dst, kernel, iterations=3)
```

```python
# cv2.RETR_EXTERNAL 
_,contours, hierarchy = cv2.findContours(dst_1, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

print(type(contours))
print(len(contours))
contours = sorted(contours, key=cv2.contourArea, reverse=True)
```

```python


center_points = []
if len(contours) > 0:
    for cont in contours:   
        box = cv2.minAreaRect(cont)
        points = np.int0(cv2.boxPoints(box))
        # 
        p1x,p1y = points[0,0], points[0,1]
        p3x,p3y = points[2,0], points[2,1]
        center =  [(p1x+p1y )/2,(p3x+p3y)/2]
        center_points.append(center)
        #draw min rectangle 
        cv2.drawContours(img_rgb_res, [points],0, (0,191,255),2)
        
    plt.figure(figsize=(10,12))
    plt.imshow(img_rgb_res)  
```

```python
if len(contours) > 0:
    #cv2.boundingRect()x, y, w, h x, yw,h
    
    boxes = [cv2.boundingRect(c) for c in contours] 
    # normal size
#     boxes_min = [cv2.minAreaRect(c) for c in contours]
    
    for box in boxes:
        
        x, y, w, h = box
        cv2.rectangle(img_res, (x, y), (x+w, y+h), (0, 191, 255), 2)
    
    plt.figure(figsize=(10,12))
    plt.imshow(img_res)
```
